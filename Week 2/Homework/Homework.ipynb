```python
#Create a dataframe that has all countries in Europe with GDP per capita less than the
#average European GDP per capita in 2007? Save your result in a variable result.

europe07avg = gap.loc[(gap.year == 2007) & (gap.continent == 'Europe'), 'gdpPercap'].mean()
#europe07 = gap.loc[(gap.year == 2007) & (gap.continent == 'Europe'), ['gdpPercap', 'country']]
#europe07
```

```python
result = pd.DataFrame(gap.loc[(gap.continent == 'Europe') & (gap.year == 2007) 
                                & (gap.gdpPercap < europe07avg),:])
result
```
