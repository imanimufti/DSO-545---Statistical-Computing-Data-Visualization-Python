```python
#24. Create the following Tree map using the mpg.csv dataset and squarify Python package.
import squarify
mpg = pd.read_csv('mpg.csv')
zyx = list(mpg['class'].unique())
```

```python
df = mpg.groupby('class').size().reset_index(name = 'counts')
#name of first column comes from column inside the groupby fn
#name of extra column created can be set from reset_index
labels = df.apply(lambda x:str(x[0]) + '\n' + '(' + str(x[1]) + ')', axis = 1)
colors = plt.get_cmap('Spectral')(np.linspace(0,1,len(labels)))
```

```python
mpg.groupby('class').size().reset_index(name = 'counts')
```

```python
plt.figure()
squarify.plot(sizes = df['counts'], 
              label = labels, 
              color = colors,
              alpha = 0.6)
plt.axis('off')
plt.show()
```
