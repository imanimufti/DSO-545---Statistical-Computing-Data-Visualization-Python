```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movie = pd.read_csv('movie_scores.csv'), index_col = 0)
movie.head()
```

```python
movie.iloc[:,1:4]
```

```python
plt.figure(figsize = (10,6))
pos = np.arange(len(movie.MovieTitle))
width = 0.3

plt.bar(pos - width/2,           plt.bar(pos + width/2,
        movie.Tomatometer,               movie.AudienceScore,
        width = width,                   width = width,
        label = 'Tomatometer')           label = 'Audience Score'

plt.xticks(pos, rotation = 10)
plt.yticks(np.arange(0,101,20))

ax = plt.gca() #get current axes for setting tick labels
ax.set_xticklabels(movie.MovieTitle)
#ax.set_yticklabels([str(i)+'%' for i in np.arange(0,101,20)])
ax.set_yticklabels(['0%', '20%', '40%', '60%', '80%', '100%'])
#adds the %age sign to the ytick labels
ax.set_yticks(np.arange(0,100,5), minor = True)
#adds minor ticks to y-axis in the interval of 5
ax.yaxis.grid(which = 'major')
#adds gridlines to the major ticks
ax.yaxis.grid(which = 'minor', linestyle = '--')
#adds gridlines to the minor ticks

plt.title('Movie Comparison')
plt.legend()
plt.show()
```
