```python
plt.figure(figsize = (11,6))

plt.plot(A, B,
        color = 'blue',
        marker = 'o',
        linestyle = '-',
        linewidth = 2,
        markersize = 12)

plt.plot(C, D,
        color = 'Red',
        marker = 's',
        linestyle = ':',
        markersize = 8,
        linewidth = 2)

plt.ylim(0,5)
plt.title('This is my first python plot',
         fontsize = 12,
         fontweight = 'bold')
plt.xlabel('X Label', fontsize = 14)
plt.ylabel('Y Label', fontsize = 14)

plt.show()
```

```python
np.random.seed(0)
```

```python
iq_scores = [int(i) for i in list(np.random.normal(100,15,100))]
```

```python
#Use the IQ scores to create the histogram
plt.figure(figsize=(6,4), dpi = 150)
plt.hist(iq_scores)

plt.axvline(x=100, color = 'r')
plt.axvline(x=85, color = 'r', linestyle = '--')
plt.axvline(x=115, color = 'r', linestyle = '--')

plt.title('IQ Scores for a test group for a hundred adults')
plt.xlabel('IQ Score')
plt.ylabel('Frequency')

plt.show()
```

```python
#Create a boxplot which shows the distribution of all IQ scores
plt.figure()
plt.boxplot(iq_scores)
plt.xlabel('Test Group')
plt.ylabel('IQ Score')
plt.title('IQ Scores for a test group for a hundered adults')
```

```python
#9. Create 4 groups of random IQ scores as follows:
np.random.seed()
group_a = [int(i) for i in list(np.random.normal(106,14,100))]
group_b = [int(i) for i in list(np.random.normal(99,14,100))]
group_c = [int(i) for i in list(np.random.normal(103,14,100))]
group_d = [int(i) for i in list(np.random.normal(97,14,100))]
```

```python
plt.figure(figsize=(6,4), dpi=150)
plt.boxplot([group_a,group_b,group_c,group_d],
           labels = ['Group A', 'Group B', 'Group C', 'Group D'])
plt.title('IQ Scores for different test groups')
plt.ylabel('IQ Score')
plt.show()
```
