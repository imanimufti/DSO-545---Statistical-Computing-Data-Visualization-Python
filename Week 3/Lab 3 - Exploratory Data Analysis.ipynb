```python
### Find the mean and std. for all x's. 
# Start by finding the mean for x1 using indexing techniques.
# to access to the column, we can use either loc or iloc

# data.iloc[ROW, COL]
# data.iloc[:,0]
# x.mean()

data.iloc[:,0].mean() #We want all the rows in x1. & then just do .mean instead of writing an extra value of code to make it faster
```

```python
# Find the mean and std. for all x's. Determine all the means for all the x's by accessing their columns.

#data.iloc[:,START:END:STEP]
data.iloc[:,::2].mean()
```
