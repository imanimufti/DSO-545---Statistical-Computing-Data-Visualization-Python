```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
trump = pd.read_csv("trump_approval.csv")
```

```python
trump
```

```python
#1
trump["Date"]= pd.to_datetime(trump["Date"]) 
```

```python
trump
```

```python
trump.info()
```

```python
#2
plt.figure(figsize = (11,6))
plt.plot("Date", "Approve", 
         data = trump, 
         label = "Approval", 
         color = "Green")
plt.plot("Date", "Disapprove", 
         data = trump, 
         label = "Disapproval", 
         color = "Orange")
plt.title("President Trump's Approval Rating", fontsize = 14, fontweight = "bold")
plt.ylabel("President's Rating (%)")         
plt.legend(loc= 'upper left')
plt.show()
```

```python
#7

plt.figure(figsize = (11,6))
plt.plot("Date", "Approve", 
         data = trump, 
         label = "Approval", 
         color = "Green")
plt.plot("Date", "Disapprove", 
         data = trump, 
         label = "Disapproval", 
         color = "Orange")

plt.title("President Trump's Approval Rating", fontsize = 14, fontweight = "bold")
plt.ylabel("President's Rating (%)")

plt.axvline(x = "2018-01-01", linestyle = '--', color = "Grey")
plt.axhline(y = 50, linestyle = ':', color = 'Red')

plt.ylim(30,70)
plt.axvspan(xmin="2018-12-12", xmax="2019-01-25", facecolor='Grey', alpha=0.3)
plt.text('2018-12-12', 65, "Dec. 22- Jan.25 \nGovernment shutdown", fontsize = 12)
plt.text("2018-01-01", 65, "One year \nin office", fontsize = 12)
plt.text("2019-08-01", 39, "39% \nApprove", fontsize = 12, color = "Green")
plt.text("2019-08-01", 57, "57% \nDispprove", fontsize = 12, color = "Orange")
plt.legend(loc= 'upper left')
plt.show()
```
