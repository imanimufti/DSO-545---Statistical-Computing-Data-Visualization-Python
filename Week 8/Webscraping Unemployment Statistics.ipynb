```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from urllib.request import urlopen
```

```python
# open the webpage
html = urlopen('https://www.bls.gov/web/laus/laumstrk.htm')
html
```

```python
## get the html code from the website
soup = BeautifulSoup(html)
soup
```

```python
table = soup.find('table', {'class':'regular'})
table
```

```python
rows = table.find_all('tr')
rows[0]
```

```python
fields = [var.text.replace('(p)',' ') for var in rows[0].find_all('th')]
fields
```

```python
Unemploymentlist = []
for i in range(1, len(rows)):
    values = [var.text.replace('\n','') for var in rows[i].find_all('td')]
    Unemploymentlist.append(values)
```

```python
Unemploymentlist
```

```python
rateList = []
rankList = []
Unemploymentlist
for i in Unemploymentlist:
    if len(i) == 2:
        rateList.append(i[0])
        rankList.append(i[1])
print(rateList)
print(rankList)
```

```python
statesList = []
for i in range(1, len(rows)):
    values1 = [var.text for var in rows[i].find_all('p')]
    if ("\n" in values1[0]) == False:
        statesList.append(values1[0])
```

```python
statesList
```

```python
df = pd.DataFrame()
```

```python
df['State'] = statesList
```

```python
df['Unemployment Rate'] = rateList
```

```python
df['Rank'] = rankList
```

```python
df['Unemployment Rate'] = pd.to_numeric(df['Unemployment Rate'])
```

```python
df['Rank'] = df['Rank'].apply(float)
```
