```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen
import re
```

```python
html = urlopen('https://en.wikipedia.org/wiki/Los_Angeles')
```

```python
soup = BeautifulSoup(html)
soup
```

```python
table = soup.find('table',{'class','wikitable collapsible'}) # mw-collapsible mw-made-collapsible
table
```

```python
headers = table.find_all('th')
headers[1:15]
```

```python
fields = [headers.get_text().replace('\n','').replace('[a]','').replace('[b]','').replace(' ','') for headers in headers][1:15]
fields
```

```python
firstcolumn = table.find_all('th')
firstcolumn = firstcolumn[15:]
```

```python
firstcol = [firstcolumn.get_text().replace('\n','').replace('[a]','').replace('[b]','') for firstcolumn in firstcolumn]
firstcol
```

```python
rows = table.find_all('tr')
rows
```

```python
df = pd.DataFrame()
for i in range(1, len(rows)):
    values = [var.text.replace('\n','') for var in rows[i].find_all('td')]
    df = df.append(pd.Series(values), ignore_index = True)
```

```python
df = df.reset_index()
```

```python
df.columns = fields
```
