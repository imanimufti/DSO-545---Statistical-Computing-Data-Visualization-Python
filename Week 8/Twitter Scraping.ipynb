```python
## Instead of doing it manually, lets generalize the definition of the Employee class

    
class Employee:
    
    raise_amount = 1.04
    
    ##constructor method ##these are all attributes
    def __init__(self, firstname, lastname, pay):
        self.firstname = firstname
        self.lastname = lastname
        self.email = firstname + '.' + lastname + '@cusc.edu'
        self.pay = pay 
        
#by default 'self' is a holder or a variable you can pass with the name of the particular object being used. 
#Self is going to come from emp_1

    def full_name(self):
        return(self.firstname + " " + self.lastname)
    
    def pay_raise(self):
        self.pay = self.pay * self.raise_amount
        
```

```python
emp_1 = Employee('Nina','Richardson', 1000000)
emp_2 = Employee('Abbass', 'Al_Sharif', 100000)
```

```python
emp_1.pay_raise()
# .pay_raise() will call the method, the method will go check def pay_raise, that will go back up to raise_amount
# and make the calculation. In order to see the new pay, we're going to call .pay as an atrribute again. 
emp_1.pay
```
