```python
### Example 3:
### Scrape the table in https://en.wikipedia.org/wiki/List_of_largest_manufacturing_companies_by_revenue
## and find the total revenue per industry
```

```python
html2 = urlopen('https://en.wikipedia.org/wiki/List_of_largest_manufacturing_companies_by_revenue')
soup2 = BeautifulSoup(html2)
soup2
```

```python
table2 = soup2.find('table',{'class':'wikitable sortable plainrowheads'})
table2
```

```python
## get all rows in the table
rows = table2.find_all('tr')
rows
```

```python
## get the variable names from the header of the table
rows[0].find_all('th')
```

```python
for var in rows[0].find_all('th'):
    print(var.text)
```

```python
cols = [var.text.replace('\n','') for var in rows[0].find_all('th')]
cols
```

```python
## extract the first row in the table (not the variables)
[var.text.replace('\n','').replace('\xa0','') for var in rows[1].find_all('td')]
```

```python
## extract all the rows in the table (not the variables)
for i in range(1, len(rows)):
    values = [var.text.replace('\n','').replace('\xa0','') for var in rows[i].find_all('td')]
    print(values)
    
    len(rows)
```

```python
## define an empty Dataframe
import pandas as pd
df = pd.DataFrame()
```

```python
for i in range(1, len(rows)):
    values = [var.text.replace('\n','').replace('\xa0','') for var in rows[i].find_all('td')]
    #print(values)
    df = df.append(pd.Series(values), ignore_index = True)
```

```python
df.head()
```

```python
df.columns = cols
df
```

```python
## make sure to remove extra formatting $signs and commas
df['Revenue (by US$ million)'] = df['Revenue (by US$ million)'].str.replace(',','')
```

```python
#change to numeric    
df['Revenue (by US$ million)'] = pd.to_numeric(df['Revenue (by US$ million)'])                                                       
```

```python
df.groupby('Industry')['Revenue (by US$ million)'].sum()
```
