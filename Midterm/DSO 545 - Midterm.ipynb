```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
recruitment = pd.read_csv('recruitment.csv')
recruitment.head()
```

```python
#1. avg attrition rate for each channel #ordered in descending
x = recruitment.groupby('recruiting_source')['attrition'].mean().reset_index()
x
```

```python
#2 new variable = 'performance_cat'. 
recruitment['performance_cat'] = recruitment.performance_rating.map({1: 'Low Performance',
                                                                    2:'Low Performance',
                                                                    3: 'High Performance',
                                                                    4:'High Performance',
                                                                    5:'High Performance'})
recruitment.head()
```

```python
#2 (continued) avg attrtition rate for both employee categories
by_performance_cat = recruitment.groupby('performance_cat').mean()
attrition_rate= by_performance_cat.iloc[:,0]
attrition_rate
```

```python
#3. Barchart = Average Sales Quota for all employees hired through different channels

applied_online = recruitment.loc[(recruitment.recruiting_source == 'Applied Online'), 'sales_quota_pct'].mean()*100
referral = recruitment.loc[(recruitment.recruiting_source == 'Referral'),'sales_quota_pct'].mean()*100
campus = recruitment.loc[(recruitment.recruiting_source == 'Campus'),'sales_quota_pct'].mean()*100
search_firm = recruitment.loc[(recruitment.recruiting_source == 'Search Firm'),'sales_quota_pct'].mean()*100

y = [search_firm, campus, referral,applied_online]
```

```python
plt.figure(figsize= (8,3))
pos = np.arange(len(y))

plt.bar(pos, y, width = 0.4)
ax = plt.gca()
ax.set_xticks(pos)
ax.set_xticklabels(['Search Firm','Campus','Referral','Applied Online'])
ax.set_yticks(np.arange(0,101,20))
ax.set_yticklabels(['0%','20%','40$','60%','80%','100%'])
plt.axhline(y = 100, linestyle = ':' , color = 'red', label = 'at 100%')
plt.legend()
plt.title('Average Sales Quota percent for each Recruting Channel')
plt.show()
```
